```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import FormatStrFormatter, MultipleLocator

plt.rcParams['font.family'] = 'MS Gothic'
#

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter, MultipleLocator

# 
dt = pd.read_csv("C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real_processed\\p_real_20231113_30.csv")

# 
column = dt[["Pressure", "Length", "Force"]]

# Pressure
first_pressure = round(column["Pressure"].iloc[0], 2)  # 

# 
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(column["Pressure"], column["Length"], column["Force"], c='r', marker='o', label=f'{first_pressure}MPa')

# 
ax.set_xlabel('Pressure(MPa)')
ax.set_ylabel('Length(mm)')
ax.set_zlabel('Force(N)')

# 
ax.set_xlim(min(column["Pressure"]), max(column["Pressure"]))
ax.set_ylim(min(column["Length"]), max(column["Length"]))
ax.set_zlim(min(column["Force"]), max(column["Force"]))

# x
ax.w_xaxis.set_major_formatter(FormatStrFormatter('%.2f'))  # 

# 
ax.legend()

# 
save_path = 'C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\graph\\real\\v1_real_single_20231115_1-1'
plt.savefig(save_path)

plt.show()


#2D


# 
column2 = dt[["Force", "Length"]]

# 
fig, ax = plt.subplots(figsize=(10, 6))

# 
ax.scatter(column2["Length"],column2["Force"], c='r', marker='o', label=f'{first_pressure}MPa')

# 
ax.set_xlabel('Length (mm)')
ax.set_ylabel('Force (N)')


# 
ax.set_xlim(min(column2["Length"]), max(column2["Length"]))
ax.set_ylim(min(column2["Force"]), max(column2["Force"]))


# 
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))  
ax.yaxis.set_major_locator(MultipleLocator(10))

# 
ax.legend()

# JPG
save_path = 'C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\graph\\real\\v1_real_single_20231115_1-2'
plt.savefig(save_path)
```
