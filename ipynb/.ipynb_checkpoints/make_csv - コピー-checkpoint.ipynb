```python
import serial
import csv
import time
import pandas as pd
from datetime import datetime

def read_sensor_and_save_data(file_count):
    file_format = "C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real\\real_{date}_{count}.csv"
    
    current_date = datetime.now().strftime("%Y%m%d")
    csv_filename = file_format.format(date=current_date, count=file_count)
    
    try:
        ser= serial.Serial('COM15', 115200, timeout=1)
        
        # 
        start_time = time.time()

        with open(csv_filename, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Time", "Length", "Pressure", "Force"])

            while True: #
                try:  # 
                    data = ser.readline().decode().strip()
                    current_time = time.time()
                    elapsed_time = int((current_time - start_time) * 1000)  # 
                    print(f"{elapsed_time} - {data}") #f-stringPythonf-string

                    if data and not data.startswith("Sensor reading stopped"):
                        parts = data.split(',') #dataparts
                        if len(parts) == 3:
                            wire_length, pressure, force = parts
                            pressure = float(pressure) / 1000.0
                            force = round(-9.80665 * float(force), 2)
                            writer.writerow([elapsed_time, wire_length, pressure, force])
                        else: #parts3
                            print("Invalid data format:", data)
                    elif data.startswith("Sensor reading stopped"):
                        print("Sensor reading stopped. Exiting loop.")
                        break #breakforwhile
                except KeyboardInterrupt: # 
                    print("Interrupt received. Saving CSV file and exiting.")
                    break

    finally: # 
        ser.close()

file_count = input("Enter file count: ")  # 
read_sensor_and_save_data(file_count)
```

```python
# 
current_date = datetime.now().strftime("%Y%m%d")
input_file_path = f"C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real\\real_{current_date}_{file_count}.csv"
df = pd.read_csv(input_file_path)


# "Length"
cleaned_df = df[df['Length'] != df['Length'].shift()]

# 
output_file_path = f"C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real_processed\\p_real_{current_date}_{file_count}.csv"
cleaned_df.to_csv(output_file_path, index=False)
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter, MultipleLocator
from mpl_toolkits.mplot3d import Axes3D
from datetime import datetime

current_date = datetime.now().strftime("%Y%m%d")
plt.rcParams['font.family'] = 'MS Gothic'  # 

# 
data_path = f"C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real_processed\\p_real_{current_date}_{file_count}.csv"
dt = pd.read_csv(data_path)

# 
column = dt[["Pressure", "Length", "Force"]]

# Pressure
first_pressure = round(column["Pressure"].iloc[0], 2)  # 

# 
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# Force
ax.plot(column["Pressure"].iloc[:column["Force"].idxmax()+1], 
        column["Length"].iloc[:column["Force"].idxmax()+1], 
        column["Force"].iloc[:column["Force"].idxmax()+1], 
        c='r',marker = 'o',markersize = 2, label=f'{first_pressure}MPa - extension')

# Force
ax.plot(column["Pressure"].iloc[column["Force"].idxmax()+1:], 
        column["Length"].iloc[column["Force"].idxmax()+1:], 
        column["Force"].iloc[column["Force"].idxmax()+1:], 
        c='b',marker = 'o',markersize = 2,label=f'{first_pressure}MPa - contraction')

# 
ax.set_xlabel('Pressure(MPa)')
ax.set_ylabel('Length(mm)')
ax.set_zlabel('Force(N)')

# x
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))  # 

# 
fig.legend(labels=['Extension', 'Contraction'], loc=(0.81,0.1), fontsize='large')

# 
save_path = f'C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\graph\\single\\single_{current_date}_{file_count}-1.jpg'
plt.savefig(save_path,bbox_inches='tight')

plt.tight_layout()  # 
plt.show()


#2D


# 
dt = pd.read_csv(f"C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real_processed\\p_real_{current_date}_{file_count}.csv")

# Pressure
column = dt[["Length", "Force"]]

# Pressure
first_pressure = round(dt["Pressure"].iloc[0], 2)  # 

# 
fig, ax = plt.subplots(figsize=(10,6))

# Force
ax.plot(column["Length"][:dt["Force"].idxmax() + 1], column["Force"][:dt["Force"].idxmax() + 1], c='r',marker = 'o',markersize = 2,label=f'{first_pressure}MPa - extension')

# Force
ax.plot(column["Length"][dt["Force"].idxmax() + 1:], column["Force"][dt["Force"].idxmax() + 1:], c='b',marker = 'o',markersize = 2,label=f'{first_pressure}MPa - contraction')

# 
ax.set_xlabel('Length (mm)')
ax.set_ylabel('Force (N)')

# 
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))

# 
ax.legend(loc = 'upper left',fontsize = 'xx-large')


# JPG
save_path = f'C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\graph\\single\\single_{current_date}_{file_count}-2.jpg'
plt.savefig(save_path,bbox_inches='tight')

# 
plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9, hspace=0.4, wspace=0.4)
plt.tight_layout()# 
plt.show()
```
