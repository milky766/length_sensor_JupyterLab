```python
import pandas as pd

# CSVデータの読み込み
df = pd.read_csv("C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real\\20231118_2_0.6\\real_20231118_2_6_0.7_s.csv")  # "your_file.csv"を実際のファイル名に変更してください

# 80個以上同じ値が連続する箇所を特定し、真ん中の値のインデックスを取得
indices_to_keep = []
current_value = None
count = 0

for i, row in df.iterrows():
    if row['Length'] == current_value:
        count += 1
    else:
        current_value = row['Length']
        count = 1

    if count >= 80:
        middle_index = i - count // 2
        indices_to_keep.append(middle_index)

# データを抽出
filtered_df = df.loc[indices_to_keep].copy()

# 80個未満の同じ値が連続する箇所を特定し、それらの行を削除
df['is_duplicate'] = df['Length'].duplicated(keep='first')
df = df[~df['is_duplicate'] | df.index.isin(indices_to_keep)]

# 不要な列を削除
df = df.drop(columns='is_duplicate')

# 結果を表示
print(filtered_df)

filtered_df.to_csv("C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real\\20231118_2_0.6\\real_20231118_2_6_0.7_ps.csv", index=False)  # "processed_data.csv"を適切なファイル名に変更してください
```

```python
import pandas as pd

# CSVデータの読み込み
df = pd.read_csv("C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real\\20231118_2_0.6\\real_20231118_2_6_0.7_s.csv")  # ファイル名は適切なものに変更してください

def process_data(df):
    result_df = pd.DataFrame(columns=df.columns)

    current_value = None
    consecutive_count = 0

    for index, row in df.iterrows():
        if row['Length'] == current_value:
            consecutive_count += 1
        else:
            consecutive_count = 1
            current_value = row['Length']

        if consecutive_count >= 100:
            middle_index = index - (consecutive_count // 2)
            result_df = pd.concat([result_df, df.loc[[middle_index]]])
            consecutive_count = 0  # 連続カウントをリセット

    return result_df

# 処理後のデータフレームを取得
processed_df = process_data(df)

# 重複行を削除
processed_df = processed_df.drop_duplicates()

# 結果の表示
print(processed_df)

filtered_df.to_csv("C:\\Users\\Mizuki\\JupyterLab\\length_sensor\\data\\real\\20231118_2_0.6\\real_20231118_2_6_0.7_ps.csv", index=False) 
```
