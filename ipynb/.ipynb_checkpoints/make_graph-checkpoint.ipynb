```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import FormatStrFormatter, MultipleLocator

plt.rcParams['font.family'] = 'MS Gothic'
#凡例やラベルに日本語を使っても文字化けしないようにする
```

```python
dt1 = pd.read_csv("sample_data\\sample_data(1).csv")
dt2 = pd.read_csv("sample_data\\sample_data(2).csv")
dt3 = pd.read_csv("sample_data\\sample_data(3).csv")
dt4 = pd.read_csv("sample_data\\sample_data(4).csv")
dt5 = pd.read_csv("sample_data\\sample_data(5).csv")

column1 = dt1[["Pressure","Length","Force"]]
column2 = dt2[["Pressure","Length","Force"]]
column3 = dt3[["Pressure","Length","Force"]]
column4 = dt4[["Pressure","Length","Force"]]
column5 = dt5[["Pressure","Length","Force"]]

combined_column = pd.concat([column1,column2,column3,column4,column5])

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection = '3d') #例えば．(221, projection='3d')(222, projection='3d')(223, projection='3d')(224, projection='3d')とすれば，4つグラフを並べられる    
ax.scatter(column1["Pressure"], column1["Force"], column1["Length"], c='r', marker='o', label='10KPa')
ax.scatter(column2["Pressure"], column2["Force"], column2["Length"], c='g', marker='o', label='20kPa')
ax.scatter(column3["Pressure"], column3["Force"], column3["Length"], c='b', marker='o', label='30kPa')
ax.scatter(column4["Pressure"], column4["Force"], column4["Length"], c='c', marker='o', label='40kPa')
ax.scatter(column5["Pressure"], column5["Force"], column5["Length"], c='m', marker='o', label='50kPa')

ax.set_xlabel('Pressure(kPa)')
ax.set_ylabel('Force(N)')
ax.set_zlabel('Length(mm)')

ax.set_xlim(min(combined_column["Pressure"]),max(combined_column["Pressure"]))
ax.set_ylim(min(combined_column["Force"]),max(combined_column["Force"]))
ax.set_zlim(min(combined_column["Length"]),max(combined_column["Length"]))

#何故かx軸が少数第一位まで表示されたので整数にフォーマット
ax.w_xaxis.set_major_formatter(FormatStrFormatter('%.0f'))  

# Pressure軸のメモリを10刻みに設定
ax.w_xaxis.set_major_locator(MultipleLocator(10))

ax.legend()

# 保存パスとファイル名を指定してグラフを画像として保存（JPG形式）
save_path = 'graph\\sample(1)-1.jpg'
plt.savefig(save_path)
```

```python
# 各データフレームの圧力を凡例用に取得
pressure1 = column1["Pressure"].iloc[0]#"Prsuure列の最初の1行の値を取得
pressure2 = column2["Pressure"].iloc[0]
pressure3 = column3["Pressure"].iloc[0]
pressure4 = column4["Pressure"].iloc[0]
pressure5 = column5["Pressure"].iloc[0]

# サブプロットの設定
fig, axes = plt.subplots(2, 3, figsize=(15, 10))#このように，２つの変数を同時に定義できる．

# タイトル yパラメータでタイトルを下に移動
#fig.suptitle("Force vs Length for Different Pressures", fontsize=20, y = 0.01) 

# 圧力に対応するカラーマップを作成. {}はディクショナリという．キーと値のペアが含まれる．キーは圧力値，値は色．
pressure_colors = {10: 'red', 20: 'green', 30: 'blue', 40: 'cyan', 50: 'magenta'} 

# 各データをプロット
data = [column1, column2, column3, column4, column5]
for i, ax in enumerate(axes.flat):#enumerate(axes.flat)は、flatによりaxesを1次元のイテレータに変換し、そのインデックス（i,0から始まる）をつずつ処理していく．
    if i < len(data):#i = 0,1,2,3,4に対して順にデータをプロット
        column = data[i]#プロットするデータをセット
        ax.scatter(column["Force"], column["Length"], c=pressure_colors[column["Pressure"].iloc[0]], marker='o')
        ax.set_xlabel('Force (N)',fontsize = 16)
        ax.set_ylabel('Length (mm)',fontsize = 16)
        ax.set_title('Pressure: {0} kPa'.format(int(column["Pressure"].iloc[0])), fontsize = 18)#{0}はフォーマット文字列内のプレースホルダで、後でその位置に挿入される値を表します。formatメソッドを使用して、{0}の位置に挿入される値を指定します。
        
        # X軸およびY軸のメモリのフォントサイズを大きくする
        ax.tick_params(axis='x', labelsize=14)
        ax.tick_params(axis='y', labelsize=14)
        
    else:
        ax.axis('off')#i = 5はプロットしない

plt.tight_layout()

#サブプロット間の間隔を調整して、図内の各サブプロットが適切に配置されるようにします。これにより、サブプロットが重なったり、テキストが重なったりすることが減少します。
#図の外側にテキストやラベルがある場合、これらの要素が切れるのを防ぎます。テキストやラベルが図からはみ出さないように配置されます。
#マージンや軸ラベルなど、図全体の要素が適切に調整され、プロットが見栄えの良い形に整形されます。

# 保存パスとファイル名を指定してグラフを画像として保存（JPG形式）
save_path = 'graph\\sample(1)-2.jpg'
plt.savefig(save_path)
```
